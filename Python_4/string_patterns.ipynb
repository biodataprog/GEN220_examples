```python
import re
import urllib.request
# got to https://github.com/GITenberg/A-Tale-of-Two-Cities_98/blob/master/book.asciidoc
url='https://raw.githubusercontent.com/GITenberg/A-Tale-of-Two-Cities_98/master/book.asciidoc'

info = urllib.request.urlopen(url)
royals = {}
allwords = {}
i = 0
for line in info:
    linestrip = line.decode('UTF-8').strip()
    if i < 5:
        print(linestrip)
    i += 1
    m = re.search(r'(king|queen|jack|ace)\s+',linestrip.lower())
    if m:
        name = m.group(1).lower()       
        if name not in royals:
            royals[name] = 0
        royals[name] += 1
    for word in linestrip.lower().split():
        if re.search(r'\w+',word):
            if word not in allwords:
                allwords[word] = 0
            allwords[word] += 1
print(royals)
print(allwords)
        
```

```python
for stmt in ('1 bird','2 birds', '20 birds', 'no birds',
            '100bird'):
    m = re.search('^(\d+)(\D*)birds?$',stmt)
    if m:
        print(stmt, ' matches')
        print(m.group(1))
        print("'{}'".format(m.group(2)))
    else:
        print(stmt, ' does not match')
```
